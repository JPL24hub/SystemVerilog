```python
from pynq import Overlay
import time
```

```python
# Load the overlay
overlay = Overlay('/home/xilinx/jupyter_notebooks/adder_demo/design_1_wrapper.bit')
```

```python
# Check available IP blocks
print("Available IP blocks:")
print(overlay.ip_dict.keys())
```

```python
# Access the adder IP directly
adder = overlay.AdderDemoIP_0

print("\n=== AXI4 Adder IP Demo ===\n")
```

```python
# Initialize values
first_value = 77
second_value = 23

# Write to registers
adder.write(0x00, first_value)
adder.write(0x04, second_value)
```

```python
# Test loop
for i in range(10):
    reg0 = adder.read(0x00)
    reg1 = adder.read(0x04)
    result = adder.read(0x08)
    
    print(f"First Value: {reg0}")
    print(f"Second Value: {reg1}")
    print(f"Sum: {result}")
    print("-" * 30)
    
    time.sleep(1)
    
    first_value += 1
    second_value += 1
    adder.write(0x00, first_value)
    adder.write(0x04, second_value)
```
